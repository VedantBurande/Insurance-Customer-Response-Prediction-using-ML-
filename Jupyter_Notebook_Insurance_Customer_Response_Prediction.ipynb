### Step 4.6 | Response Rate By Age Group 

```python
# Convert Age column to numeric, handle NaN
data["Age"] = pd.to_numeric(data["Age"], errors="coerce")
data["Age"].fillna(data["Age"].median(), inplace=True)  # Replace NaN with median

# Define Age bins and labels
age_bins = [0, 18, 25, 35, 45, 55, 100]  # Added 100 to avoid errors for ages > 55
age_labels = ['<18', '18-24', '25-34', '35-44', '45-55', '55+']

# Create Age Group column
data["AgeGroup"] = pd.cut(data["Age"], bins=age_bins, labels=age_labels, include_lowest=True)

# Check and handle NaN in Response
data["Response"].fillna(0, inplace=True)

# Group by Age Group and calculate mean Response Rate
age_conversion = data.groupby("AgeGroup")["Response"].mean().reset_index()

# Plot Bar Chart
plt.figure(figsize=(8, 5))
sns.barplot(x="AgeGroup", y="Response", data=age_conversion, palette="coolwarm")
plt.xlabel("Age Group")
plt.ylabel("Average Response Rate")
plt.title("Response Rate by Age Group")
plt.show()
```

### Step 6.1 | Base Models

#### Step 6.1.1 | Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)

y_pred_log = log_reg.predict(X_test)

# Generate classification report
print("\nClassification Report:\n", classification_report(y_test, y_pred_log))

# Compute confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred_log)
print("\nConfusion Matrix:\n", conf_matrix)

# Compute accuracy score
accuracy = accuracy_score(y_test, y_pred_log)
print("\nAccuracy Score:", accuracy)
```

#### Step 6.1.2 | Random Forest
